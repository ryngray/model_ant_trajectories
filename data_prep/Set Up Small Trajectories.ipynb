```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = pd.read_csv("ants.txt", delimiter=',')
```

```python
print(f.keys())
print(f.head())
print(np.unique(f['id']))
```

```python
d1 = f[f['id']==1]
```

```python
small_traj_id = np.ones(np.shape(f)[0])*(-1)

f['sub_traj_id'] = small_traj_id
```

```python
sub_traj = 0
epsilon = 100
traj_total = []
for i in np.unique(f['id']):
    print(i)
    fi = f[f['id']==i]
    tmp_list = []
    for j in range(np.shape(fi)[0]):
        if(j==0):
            tmp_list.append(fi.iloc[j])
        elif(abs(fi.iloc[j]['x'] - fi.iloc[j-1]['x']) < epsilon and abs(fi.iloc[j]['y'] - fi.iloc[j-1]['y']) < epsilon):
            tmp_list.append(fi.iloc[j])
        else:
            #print("ERROR!", fi.iloc[j]['x'], fi.iloc[j-1]['x'])
            tmp_list = []
        if(len(tmp_list) == 1500):
            traj_total.append(tmp_list)
            tmp_list = []
    print(np.shape(traj_total))
    sub_traj+=1
```

```python
print(np.shape(traj_total))
traj_total = np.asarray(traj_total)
```

```python
arr_reshaped = traj_total.reshape(traj_total.shape[0], -1)
  
# saving reshaped array to file.
np.savetxt("trajectories_split.txt", arr_reshaped)
  
# retrieving data from file.
loaded_arr = np.loadtxt("trajectories_split.txt")
  
# This loadedArr is a 2D array, therefore
# we need to convert it to the original
# array shape.reshaping to get original
# matrice with original shape.
load_original_arr = loaded_arr.reshape(
    loaded_arr.shape[0], loaded_arr.shape[1] // traj_total.shape[2], traj_total.shape[2])
  
# check the shapes:
print("shape of arr: ", traj_total.shape)
print("shape of load_original_arr: ", load_original_arr.shape)
  
# check if both arrays are same or not:
if (load_original_arr == traj_total).all():
    print("Yes, both the arrays are same")
else:
    print("No, both the arrays are not same")
```
